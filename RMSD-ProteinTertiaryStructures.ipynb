<h2><center>Abstract</center></h2>
<div style="text-align:justify">
One of the most important lines of research in the field of molecular biology is the three-dimensional reconstruction of macromolecules like proteins. The prediction of protein structure is one of the most important challenges in bioinformatics. Correctly predicting only the spinal column of a tertiary structure presents a great difficulty to determinate the flexible torsion angles, as well as angles of the individual side chains. There are an average of hundreds of variables in total for only a protein of moderate size. For this reason, for some years the soft computing technologies are emerging as good candidates for the solution of this type of NP-complete problems. This research presents an exploratory analysis and the prediction of the RMSD (Root Mean Square Deviation) of a decoy set with size ranging from zero to twenty one Ångströms. Finally, we corroborated what was reported in the literature, one of the best regression algorithms for this data set is the Random Forest, in which case an root-mean-square error of 3.75  was obtained after the processing presented here.
</div>

<h2><left>1. Problem Statement</left></h2>
<div style="text-align:justify">
Proteins are biological polymers composed of amino acids. A polypeptide chain is formed by amino acids linked together by peptide bonds and it's known as primary structure. One or more polypeptide chains twisted in a 3-D shape form a protein secondary structure. <i><b>Proteins have complex shapes that include multiple folds, loops and curves are considered polypeptide chain and are known as tertiary structures.</b></i> A protein macromolecule formed by interactions between multiple polypeptide chains, i.e, sets of tertiary structures are called protein quaternary structure[1]. Figure 2 shows the different kinds of protein structures, their shapes and complex compositions. Protein Tertiary structure is so important, e.g. in enzyme functionalities since they spatially connect all the functional pieces of a protein together to perform a task. Usually, this task is so specific in relation to that level of structure.<br>
<br>

A protein consists of a polypeptide backbone with attached side chains how Figure 2 evidence it. Each type of protein differs in its number of amino acids and its sequence; thus, it is the sequence of the chemically different side chains that makes each protein distinct. When two or more amino acids combine to form a peptide, the elements of water are removed, and what remains of each amino acid is called an amino-acid residue. When two or more amino acids combine to form a peptide, the elements of water are removed, and what remains of each amino acid is called an amino-acid residue. <i><b>The residue can originate from amino acids or peptides and generally have molecular groups amino, carboxyl or hydroxyl in free state and they can be linked by different bonds to other molecules[2].</b></i> Figure 2 show circles that represent different kinds of amino acids that can be polar or non-polar side chains. The non-polar amino acid side chains are buried on the inside forming a tightly packed hydrophobic core of atoms that are hidden from water. <i><b>Hydrophobic interactions are important for the folding of proteins. This is important in keeping a protein stable and biologically active, because it allow to the protein to decrease in surface are and reduce the undesirable interactions with water[3]</b></i>.<br>
<br>
<center><img src="images/hidrophobic_effect.png", width='80%'></center>
<font size="-1"><center>
<b>Figure 2:</b>Tertiary structure and hydrophobic effect. Source: [1]</center></font><br>

The interaction of the neighboring parts of the polypeptide chain can restrict the access of water molecules to the ligand binding sites of that protein. Because water molecules easily form hydrogen bonds that can compete with ligands for sites on the surface of the protein, a ligand forms narrower hydrogen bonds (and electrostatic interactions) with a protein if the water molecules are maintained remote. <i><b>Inherently, the lower the exposure of the non-polar amino acids and their respective residues, the lower the hydrophobic core exposure of the protein tertiary structure.</b></i> Indeed, a protein can keep a site that binds to the ligand dry, increasing the site's reactivity, because it is energetically unfavorable for the individual water molecules to separate it due to the strong tendency of water molecules to form water–water hydrogen bonds, as they must to achieve a crack on the surface of a protein. <i><b>Consequently, the protein will have greater stability and the quality of the protein could be evaluated depending on this condition. In other words, the surface conformation of a protein tertiary structures determines its chemistry[1]</b></i>.<br>

```python
def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
train_set, test_set = split_train_test(df, 0.2)
print(len(train_set), "train +", len(test_set), "test")
```

```python
def test_set_check(identifier, test_ratio):
    return crc32(np.int64(identifier)) & 0xffffffff < test_ratio * 2**32

def split_train_test_by_id(data, test_ratio, id_column):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio))
    return data.loc[~in_test_set], data.loc[in_test_set]
```

```python
def test_set_check(identifier, test_ratio, hash=hashlib.md5):
    return hash(np.int64(identifier)).digest()[-1] < 256 * test_ratio
```

<div style="text-align:justify"> 
Now, for the transformation of characteristics, the class <code> CombinedAttributesAdder()</code> is created, which will allow to add attributes obtained from internal operations within a pipeline.
</div>

```python
F1_ix, F2_ix, F3_ix, F4_ix, F5_ix, F6_ix = 1, 2, 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_F10 = False):
        self.add_F10 = add_F10
    def fit(self, X, y=None):
        return self  
    def transform(self, X, y=None):
        F11 = X[:, F3_ix] / X[:, F5_ix]
        F12 = X[:, F6_ix] / X[:, F5_ix]
        if self.add_F10:
            F10 = X[:, F2_ix] / X[:, F1_ix]
            return np.c_[X, F11, F12, F10]
        else:
            return np.c_[X, F11, F12]

attr_adder = CombinedAttributesAdder(add_F10=False)
df_extra_attribs = attr_adder.transform(df.values)
```

```python
df_extra_attribs = pd.DataFrame(df_extra_attribs,columns=list(df.columns)+["F11", "F12"])
df_extra_attribs.head()
```

<div style="text-align:justify"> 
We define now with the <code>Pipeline</code> module of Scikit-Learn library a workflow that performs an <code>Imputer</code> under the criterion of mean value, with which all the missing values of a given caracetristica assigns it the mean value of the entire population. Then he executes the class for the combination of attributes that we previously defined and finally applies a standard scaler that subtracts the average value element-wise to each element of the population and maintains a variance $\sigma_{f}^{2}=1$ at the set.
</div>

```python
num_pipeline = Pipeline([('imputer', Imputer(strategy="median")),('attribs_adder', CombinedAttributesAdder()),
                         ('std_scaler', StandardScaler()),])

df_num_tr = num_pipeline.fit_transform(df)
print(df_num_tr)
```

<h3><left>Other Methodology: My Own Pipeline and Tuned Regressors</left></h3>
<br>
<div style="text-align:justify"> 
For the proposed pipeline of this investigation, after analyzing the characteristics of the dataset, it is proposed to eliminate combining the characteristics <b>F3/F5 (Fractional area of exposed non polar residue / Molecular mass weighted exposed area)</b>, <b>F4/F5 (Fractional area of exposed non polar part of residue / Molecular mass weighted exposed area)</b> and <b>F4/F6 (Fractional area of exposed non polar part of residue / Average deviation from standard exposed area of residue)</b>, based on the explanation of the description and the close relationship between the prlofiracion of the exposed areas of the non-polar amino acids and their respective residues, with the intessability that they produce in the protein and that is manifested in the <b>RMSD</b> of the macromolecule. In addition, after combining <b>F5</b>, it is eliminated from the training dataset because of its proportional relationship with <b>F1</b> that would produce an overfitting effect.
</div>


```python
# column index
F1_ix, F2_ix, F3_ix, F4_ix, F5_ix, F6_ix = 1, 2, 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_F10 = False): # no *args or **kargs
        self.add_F10 = add_F10
    def fit(self, X, y=None):
        return self  # nothing else to do
    def transform(self, X, y=None):
        F11 = X[:, F3_ix] / X[:, F5_ix]
        F12 = X[:, F4_ix] / X[:, F5_ix]
        F13 = X[:, F4_ix] / X[:, F6_ix]
        X = np.delete(X, 0, 1)
        X = np.delete(X, 5, 1)
        if self.add_F10:
            F10 = X[:, F2_ix] / X[:, F1_ix]
            return np.c_[X, F11, F12, F13, F10]
        else:
            return np.c_[X, F11, F12, F13]

attr_adder = CombinedAttributesAdder(add_F10=False)
df_extra_attribs = attr_adder.transform(df.values)

num_pipeline = Pipeline([('imputer', Imputer(strategy="median")),('attribs_adder', CombinedAttributesAdder()),
                         ('std_scaler', StandardScaler()),])
```
